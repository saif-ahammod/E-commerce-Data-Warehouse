## 3.1: Q1: Find the division/district/year/month wise total_sale_price joining fact table and respective dimension table 

```python
con = PostgresConnection().getConnection()
queries = """
            SELECT s.division, s.district, tim.year, tim.month, SUM(CAST(t.total_price AS DECIMAL))
            FROM ecomdb.fact_table t
                JOIN ecomdb.store_dim s on s.store_key=t.store_key
                JOIN ecomdb.time_dim tim on tim.time_key=t.time_key
            GROUP BY ROLLUP (s.division,s.district,tim.year, tim.month)
                ORDER BY s.division """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(queries)
# fetch the results
records = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
print(records)
```

## 3.2: Q2: Find the customer/bank/transaction(cash/online) wise total_sale_price joining fact table and respective dimension table 

```python
con = PostgresConnection().getConnection()
query = """
            SELECT
                trans_dim.bank_name AS bank_name,
                coustomer_dim.name AS customer_name,
                trans_dim.trans_type AS transaction_type,
                SUM(CAST(fact_table.total_price AS INTEGER)) AS total_sale_price
            FROM
                ecomdb.fact_table
                JOIN ecomdb.trans_dim ON fact_table.payment_key = trans_dim.payment_key
                JOIN ecomdb.coustomer_dim ON fact_table.coustomer_key = coustomer_dim.coustomer_key
            GROUP BY
                ROLLUP(bank_name, customer_name, transaction_type)
            """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(query)
# fetch the results
record = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
df = pd.DataFrame(record, columns=['bank_name', 'coustomer_name', 'transaction_type', 'total_sale_price'])
print(df)
```

## 3.3: Q3: Total sales in Barisal for item 'Pepsi - 12 oz cans'

```python
con = PostgresConnection().getConnection()
queries = """
            SELECT st.district, SUM(CAST(ft.total_price AS DECIMAL))
                FROM ecomdb.fact_table as ft
                JOIN ecomdb.store_dim as st on st.store_key=ft.store_key
                JOIN ecomdb.item_dim as it on it.item_key = ft.item_key
                WHERE st.district = 'BARISAL' and it.item_name = 'Pepsi - 12 oz cans'
                GROUP BY st.district
            """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(queries)
# fetch the results
records = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
print(records)
```

## 3.6: Q6: For each store(item supplier), what are the top three products offered that are most often purchased?
S1-> item1 -> quantity_sales

S1-> item2 -> quantity_sales

S1-> item3 -> quantity_sales

```python
con = PostgresConnection().getConnection()
query = """
            SELECT st.store_key AS Store, 
                   it.item_name AS Item, 
                   SUM(ft.quantity) AS "Quantity Sold"
            FROM ecomdb.fact_table AS ft
            JOIN ecomdb.store_dim st ON ft.store_key = st.store_key
            JOIN ecomdb.item_dim it ON ft.item_key = it.item_key
            GROUP BY Store, Item
            ORDER BY Store, "Quantity Sold" DESC
            LIMIT 3;
            """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(query)
# fetch the results
record = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
df = pd.DataFrame(record, columns=['Store', 'Item', 'Quantity Sold'])
print(df)
```

## 3.8: Q8: What season(quarter) is the worst for each product item? As example,

item1-> q1

item2-> q2

```python
con = PostgresConnection().getConnection()
query = """
            SELECT it.item_name, tim.quarter,
                SUM(ft.quantity) AS quantity_sale
            FROM ecomdb.fact_table ft
            JOIN ecomdb.item_dim it ON ft.item_key = it.item_key
            JOIN ecomdb.time_dim tim ON ft.time_key = tim.time_key
            GROUP BY ROLLUP(it.item_name, tim.quarter)
            HAVING it.item_name IS NOT NULL AND tim.quarter IS NOT NULL
            ORDER BY it.item_name, quantity_sale ASC
            """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(query)
# fetch the results
record = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
df = pd.DataFrame(record, columns=['Item', 'Quarter', 'Quantity Sales'])

 # Create a dictionary to store the worst quarter for each product
worst_quarters = {}

# Loop through each item in the dataframe
for index, row in df.iterrows():
    item = row['Item']
    quarter = row['Quarter']
    sales = row['Quantity Sales']

    # If the item is not in the dictionary yet, add it with the current quarter as the worst
    if item not in worst_quarters:
        worst_quarters[item] = quarter

    # If the item is already in the dictionary, update the worst quarter if the current sales is lower
    else:
        if sales < df.loc[(df['Item'] == item) & (df['Quarter'] == worst_quarters[item]), 'Quantity Sales'].values[0]:
            worst_quarters[item] = quarter

# Create a new dataframe to store the worst quarters for each product
worst_quarters_df = pd.DataFrame(worst_quarters.items(), columns=['Item', 'Worst Quarter'])
print(worst_quarters_df)
```

## 3.9: Q9: Break down the total sales of items geographically (division-wise).

item1-> division1->total_sales

item1-> division2->total_sales

```python
con = PostgresConnection().getConnection()
query = """
            SELECT it.item_name, st.division, SUM(ft.total_price) AS total_sales
            FROM ecomdb.fact_table ft
            JOIN ecomdb.item_dim it ON ft.item_key = it.item_key
            JOIN ecomdb.store_dim st ON ft.store_key = st.store_key
            GROUP BY ROLLUP(it.item_name, st.division)
            HAVING it.item_name IS NOT NULL AND st.division IS NOT NULL
            ORDER BY it.item_name, SUM(ft.total_price) DESC
            """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(query)
# fetch the results
record = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
df = pd.DataFrame(record, columns=['Item', 'Division', 'Total Sales'])
print(df)
```

## 3.10: Q10: What are the average sales of products sales per store monthly?

S1->M1-> avg_sales

s1-> M2 -> avg_sales

```python
con = PostgresConnection().getConnection()
query = """
        SELECT st.store_key, td.month, AVG(ft.total_price) AS avg_sales
        FROM ecomdb.fact_table ft
        JOIN ecomdb.store_dim st ON ft.store_key = st.store_key
        JOIN ecomdb.time_dim td ON ft.time_key = td.time_key
        GROUP BY ROLLUP(st.store_key, td.month)
        HAVING st.store_key IS NOT NULL AND td.month IS NOT NULL
        ORDER BY st.store_key, td.month
        """
cur = con.cursor()
cur.execute("BEGIN")
cur.execute(query)
# fetch the results
record = cur.fetchall()
# commit the transaction
con.commit()
# close the cursor
cur.close()
df = pd.DataFrame(record, columns=['Store', 'Month', 'Avg Sales'])
print(df)
```
